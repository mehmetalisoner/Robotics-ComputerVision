```python
# Inputting src point coordinates (bottom left, bottom right, top left and top right corner)
import skimage
from skimage.transform import warp
from google.colab.patches import cv2_imshow


dest_points  = np.array([
    [130,   550],
    [350,    520],
    [110,230],
    [330,220]])
src_points= np.array([
    [110,    350],
    [290,    470],
    [100,240],
    [300,240]])








def create_rows_for_H(src_points, dest_points, i):
    row1 = np.array([
                    [ -src_points[i][0],-src_points[i][1],-1,0,0,0,
                    (dest_points[i][0]*src_points[i][0]),
                    (dest_points[i][0]*src_points[i][1]),                    
                    dest_points[i][0] ] 
                    ])
    row2 = np.array( [ 
                      [0,0,0,-src_points[i][0],-src_points[i][1],-1,
                     dest_points[i][1]*src_points[i][0],
                     dest_points[i][1]*src_points[i][1],
                     dest_points[i][1] ]
                      ])
    

    return np.row_stack((row1,row2))


temp = create_rows_for_H(src_points,dest_points,0)
temp2 = create_rows_for_H(src_points,dest_points,1)
temp3 = create_rows_for_H(src_points,dest_points,2)
temp4 = create_rows_for_H(src_points,dest_points,3)

temp = np.row_stack((temp,temp2,temp3,temp4))

u, s, vh = np.linalg.svd(temp)
u.shape, s.shape, vh.shape



H_matrix = []
temp = []
last_column = vh[-1,:]
for i in range(3):
  for j in range(3):
    temp.append(last_column[(i*3)+j])
  H_matrix.append(temp)
  temp = []




H_matrix = np.array(H_matrix)
print("H = ", H_matrix)

#identity=np.array([[1,0,0],[0,1,0],[0,0,1]])



warped = warp(img_dest, H_matrix)
plt.imshow(warped,cmap='gray')
plt.show()

```
