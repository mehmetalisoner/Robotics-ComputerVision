```python
# Problem 1
import numpy as np
from numpy import linalg
import scipy as scp
from scipy.linalg import svd

import matplotlib.pyplot as plt

from matplotlib.lines import Line2D 

# Defining parameters and the camera matrix

R_W = np.array([[0.707,0.707,0],[-0.707,0.707,0],[0,0,1]])
t_w = np.array([-5,0.5,4])
t_w = t_w.reshape((-1,1))
K = np.array([[-100,0,200],[-0,-100,200],[0,0,1]])
T_W = np.hstack([R_W,t_w])
M = np.dot(K,T_W)


# test = np.array([[1,0,0],[0,1,0],[0,0,1]])
# test2 = np.array([0,0,1])
# test2 = test2.reshape((-1,1))
# test3 = np.hstack([test,test2])
# M = np.dot(K,test3)

# Creating a cube, defining the points in world-coord.
p0 = np.array([0,0,0,1])
p1 = np.array([1,0,0,1])
p2 = np.array([0,0,1,1])
p3 = np.array([1,0,1,1])
p4 = np.array([0,1,1,1])
p5 = np.array([1,1,1,1])
p6 = np.array([0,1,0,1])
p7 = np.array([1,1,0,1])

points_3d = [p0,p1,p2,p3,p4,p5,p6,p7]



# Going from 3d to 2d, to image coords.
all_xs_2d= []  # store all 2D x-coordinates here
all_ys_2d= []  # store all 2D y-coordinates here

for i in range(len(points_3d)):
    temp_p = np.dot(M,points_3d[i].T)
    all_xs_2d.append(temp_p[0]/temp_p[2])   # get x-value and divide it by z-value
    all_ys_2d.append(temp_p[1]/temp_p[2])   # get y-value and divide it by z-value

# Plot using 2DLine
print(all_xs_2d)
print(all_ys_2d)
line0 = Line2D([all_xs_2d[0],all_xs_2d[1]],[all_ys_2d[0],all_ys_2d[1]])
line1 = Line2D([all_xs_2d[0],all_xs_2d[2]],[all_ys_2d[0],all_ys_2d[2]])
line2 = Line2D([all_xs_2d[1],all_xs_2d[3]],[all_ys_2d[1],all_ys_2d[3]])
line3 = Line2D([all_xs_2d[2],all_xs_2d[3]],[all_ys_2d[2],all_ys_2d[3]])
line4 = Line2D([all_xs_2d[2],all_xs_2d[4]],[all_ys_2d[2],all_ys_2d[4]])
line5 = Line2D([all_xs_2d[3],all_xs_2d[5]],[all_ys_2d[3],all_ys_2d[5]])
line6 = Line2D([all_xs_2d[4],all_xs_2d[5]],[all_ys_2d[4],all_ys_2d[5]])
line7 = Line2D([all_xs_2d[1],all_xs_2d[7]],[all_ys_2d[1],all_ys_2d[7]])
line8 = Line2D([all_xs_2d[0],all_xs_2d[6]],[all_ys_2d[0],all_ys_2d[6]])
line9 = Line2D([all_xs_2d[6],all_xs_2d[7]],[all_ys_2d[6],all_ys_2d[7]])
line10 = Line2D([all_xs_2d[4],all_xs_2d[6]],[all_ys_2d[4],all_ys_2d[6]])
line11 = Line2D([all_xs_2d[5],all_xs_2d[7]],[all_ys_2d[5],all_ys_2d[7]])













fig = plt.figure()
ax = fig.add_subplot(111)
ax.add_line(line0)
ax.add_line(line1)
ax.add_line(line2)
ax.add_line(line3)
ax.add_line(line4)
ax.add_line(line5)
ax.add_line(line6)
ax.add_line(line7)
ax.add_line(line8)
ax.add_line(line9)
ax.add_line(line10)
ax.add_line(line11)

ax.set_xlim(min(all_xs_2d), max(all_xs_2d))
ax.set_ylim(min(all_ys_2d), max(all_ys_2d))



plt.show()





```
